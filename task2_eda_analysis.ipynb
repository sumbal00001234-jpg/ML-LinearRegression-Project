```python
# Task 2 - EDA (Exploratory Data Analysis)
# Simple and clean code

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
print("="*50)
print("STEP 1: Loading dataset")
print("="*50)

df = pd.read_csv('heart.csv')
print(f"Dataset loaded!")
print(f"Shape: {df.shape} rows and columns")
print(f"Columns: {list(df.columns)}")
print()
```

```python
# Check missing values
print("="*50)
print("STEP 2: Checking for missing values")
print("="*50)

print(df.isnull().sum())
print("\nTotal missing values:", df.isnull().sum().sum())

if df.isnull().sum().sum() == 0:
    print("✅ No missing values! Good to go.")
else:
    print("⚠️ We need to handle missing values")
print()
```

```python
# Statistical summary
print("="*50)
print("STEP 3: Statistical Summary")
print("="*50)

print(df.describe())
print()
```

```python
# Visualizations
print("="*50)
print("STEP 4: Creating Visualizations")
print("="*50)

# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))

# 4.1: Histograms (Distribution of numerical features)
print("Creating histograms...")
df.hist(figsize=(15, 10), bins=20)
plt.suptitle("Histograms of All Features", fontsize=16)
plt.tight_layout()
plt.savefig('histograms.png')
plt.show()

# 4.2: Correlation heatmap
print("\nCreating correlation heatmap...")
plt.figure(figsize=(12, 8))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap", fontsize=16)
plt.tight_layout()
plt.savefig('heatmap.png')
plt.show()

# 4.3: Pairplot (only first 5 columns to avoid overcrowding)
print("\nCreating pairplot (this may take a moment)...")
sns.pairplot(df.iloc[:, :5])  # First 5 columns only
plt.suptitle("Pairplot of First 5 Features", fontsize=16)
plt.savefig('pairplot.png')
plt.show()

# 4.4: Boxplots for outlier detection
print("\nCreating boxplots...")
fig, axes = plt.subplots(3, 4, figsize=(15, 10))
axes = axes.ravel()

for i, col in enumerate(df.columns[:11]):  # First 11 columns
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(col)

plt.suptitle("Boxplots for Outlier Detection", fontsize=16)
plt.tight_layout()
plt.savefig('boxplots.png')
plt.savefig('boxplots.png')
plt.show()

print("\n✅ EDA Complete! All visualizations saved.")
```
